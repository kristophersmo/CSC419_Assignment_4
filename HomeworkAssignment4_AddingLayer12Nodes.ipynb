```python
# CSC 419 Homework Assignment 4
# Introduction to Neural Networks
# Adding Hidden Layer of 12 Nodes

# Code run/edit by Kristopher Smolarek (F00568254)
# Submitted June 5, 2024

import numpy as np  # to load data set
import torch  # to build the neural network
import torch.nn as nn
import torch.optim as optim
```

```python
# Load .csv ataset
dataset = np.loadtxt('pima_indians_diabetes.csv', delimiter=',')

# Split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# Convert data set to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).reshape(-1, 1)
```

```python
# Defining the neural network model
model = nn.Sequential(
    nn.Linear(8, 12),
    nn.ReLU(),
    nn.Linear(12, 12),
    nn.ReLU(),
    nn.Linear(12, 12),
    nn.ReLU(),
    nn.Linear(12, 1),
    nn.Sigmoid())

print(model)
```

```python
# Setting the loss function and optimizer
loss_fn = nn.BCELoss()  # binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training the neural network model
n_epochs = 100
batch_size = 10

for epoch in range(n_epochs):
    for i in range(0, len(X), batch_size):
        Xbatch = X[i:i+batch_size]
        y_pred = model(Xbatch)
        ybatch = y[i:i+batch_size]
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Finished epoch {epoch}, latest loss {loss}')
```

```python
# Model Evaluation (determining accuracy)
with torch.no_grad():
    y_pred = model(X)

accuracy = (y_pred.round() == y).float().mean()
print(f"Accuracy {accuracy}")
```
